````python
# chatbot full code

import numpy as np
import matplotlib.pyplot as plt

# Given data
sample_size = 80
successes = 49
observed_proportion = successes / sample_size

# Number of bootstrap simulations
num_simulations = 10000

# Simulating the bootstrapped distribution
bootstrap_proportions = np.zeros(num_simulations)

# Perform bootstrap sampling
for i in range(num_simulations):
    # Simulate a sample of size 80 with success probability as observed proportion
    bootstrap_sample = np.random.choice([0, 1], size=sample_size, p=[1-observed_proportion, observed_proportion])
    bootstrap_proportions[i] = np.mean(bootstrap_sample)

# Plot the bootstrapped distribution of proportions
plt.hist(bootstrap_proportions, bins=30, color='skyblue', edgecolor='black', density=True)
plt.axvline(observed_proportion, color='red', linestyle='dashed', linewidth=2, label=f"Observed Proportion ({observed_proportion:.2f})")
plt.title('Bootstrapped Distribution of Proportion of Successes')
plt.xlabel('Proportion of Successes')
plt.ylabel('Density')
plt.legend()
plt.show()

# Optional: calculate confidence interval (e.g., 95%)
lower_bound = np.percentile(bootstrap_proportions, 2.5)
upper_bound = np.percentile(bootstrap_proportions, 97.5)

print(f"95% Confidence Interval: [{lower_bound:.3f}, {upper_bound:.3f}]")
````
